```python
# GERADOR DE SENHAS P/ ARQUIVO TXT-------------------------------------------------------------------------
# 
import random
import string
import datetime
import os
# 
caracteres = string.ascii_uppercase + string.ascii_lowercase + string.digits
# 
print('GERADOR DE SENHAS P/ ARQUIVO TXT')
print('-'*50)
# 
def gerador():
    with open ('senha.txt', 'a') as arquivo:
        while True:
            try:
                tamanho_senha = input('DEFINA O TAMANHO DA QUANT. DE CARACTERES DESEJADO EM SUA SENHA (ou pressione "esc" para sair):')
                if tamanho_senha == '':
                    raise KeyboardInterrupt
                senha = ''.join(random.choices(caracteres, k = int(tamanho_senha)))
                print(f'Senha gerada: {senha} \n' + '-'*50)
                arquivo.write(senha + '\n')
            except KeyboardInterrupt:
                print('PROGRAMA ENCERRADO' + '\n' + '-'*50)
                break

# gerando arquivo txt e iniciando programa------------------------------------------------------------------
with open ('senha.txt', 'r') as arquivo:
    arquivo.read()
with open ('senha.txt', 'a') as arquivo:
    now = datetime.datetime.now()
    string_horario = now.strftime('%d/%m/%Y | %H:%M')
    if os.stat('senha.txt').st_size > 0:
        arquivo.write('\n')
    arquivo.write(f'SENHAS GERADAS POR "gerador_senha.ipynb" em {string_horario}' + '-'*50 + '\n')
gerador()

# imprimindo em terminal arquivo txt gerado-----------------------------------------------------------------
with open ('senha.txt', 'r') as arquivo:
    conteudo = arquivo.read()
print(conteudo)
```
